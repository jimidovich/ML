```python
kal = pd.read_csv('./df_kfres_1e5/ag_au.csv')
```

```python
kal
```

```python
yname, xname = kal.columns[1:3]
```

```python
kal['ez_thresh'] = abs(kal.ez.rolling(600).quantile(0.999))
kal['ypos'] = kal.apply(lambda x: -np.sign(x['ez']) if abs(x['ez']) > abs(x['ez_thresh']) else 0, axis=1) * 100
kal['xpos'] = round(-kal['ypos'] * kal['beta'])
```

```python
kal['ycomm'] = abs(kal.ypos.diff()) * kal[yname] * fut_info.ix[yname, 'comm/val%%'] / 1e4
kal['xcomm'] = abs(kal.xpos.diff()) * kal[xname] * fut_info.ix[xname, 'comm/val%%'] / 1e4
kal['pnl'] = kal[yname].diff() * kal.ypos.shift(1) + kal[xname].diff() * kal.xpos.shift(1)
kal['netpnl'] = kal.pnl - kal.ycomm - kal.xcomm
kal['equity'] = kal.netpnl.cumsum()
```

```python
kal['no_trade'] = kal.apply(lambda x: 1 if x['ycomm'] != 0 else 0, axis=1).cumsum() / 2
```

```python
sns.set_style('dark')
kal.equity.plot()
kal.pnl.cumsum().plot()
(kal.ycomm + kal.xcomm).cumsum().plot(label='commission')
plt.legend(loc=0)
kal.no_trade.plot(secondary_y=True)
```

```python
kal['cap'] = (kal[yname].max() * fut_info.ix[yname, 'margin'] + 
              kal[xname].max() * fut_info.ix[xname, 'margin']/100 * kal.xpos.max()) + kal.equity
```

```python
kal['ret'] = np.log(kal.cap).diff()
```

```python
sharpe = kal.ret.mean() / kal.ret.std() * np.sqrt(252*530)
```
